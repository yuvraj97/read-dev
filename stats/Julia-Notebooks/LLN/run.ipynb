# [Law of Large Numbers](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln) Simulation
Say we have $n$ **i.i.d** random variable $X_1,X_2,....,X_n$ with some distribution say $\mathbb{P}$  
$$\overline{X}_n:=\frac{1}{n}\sum _{i=1}^ n X_ i \xrightarrow [n\to \infty ]{\mathbb{P},\text{ a.s.}} \mu$$
$:=\quad$ this symbol means <b>"by definition"</b><br>
$\mathbb{P}\quad$ it means <b>"in Probability"</b><br>
$\text{a.s.}\quad$ it means <b>"almost surely"</b><br>
Then [LLN](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln) says,<br>
$$\left|\overline{X}_n -\mu \right| \propto \frac{1}{f(n)}$$
where $f(n)$ is an increasing function <span title="with respect to"><u>w.r.t.</u></span> \(n\).<br>
So $$\left|\overline{X}_n -\mu \right| \xrightarrow [n\to \infty ]{\mathbb{P},\text{ a.s.}} 0\$$  

- [Normal Distribution $X_i\sim\mathcal{N}(\mu,\sigma)$](#$X_i\sim\mathcal{N}(\mu,\sigma)$)

- [Bernoulli Distribution $X_i\sim\text{Ber}(p)$](#$X_i\sim-\text{Ber}(p)$)
- [Geometric Distribution $X_i\sim\text{Geo}(p)$](#$X_i\sim-\text{Geo}(p)$)

### $X_i\sim\mathcal{N}(\mu,\sigma)$

Now assume that our **sample** $X_1,X_2,....,X_n$ is drawn from a Normal Distribution, $X_i\sim\mathcal{N}(\mu,\sigma)$

  where $\sigma$ is standard deviation.

Now let's check how [LLN](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln)
behaves for Normal Distribution, $\mathcal{N}(\mu,\sigma)$   
Our expectation is $$\left|\overline{X}_n -\mu \right| \xrightarrow [n\to \infty ]{\mathbb{P},\text{ a.s.}} 0$$

### $X_i\sim \text{Exp}(\lambda)$ 
Now assume that our **sample** $X_1,X_2,....,X_n$ is drawn from a Exponential Distribution, $X_i\sim \text{Exp}(\lambda)$  
pdf of $X_i$ is $$f(X=x)=\lambda e^{-\lambda x}$$  

> in [Julia](https://juliastats.org/) Exponential Distribution is denoted as $\text{Exp}(\theta)$ and $\theta=\frac{1}{\lambda}$   
  So pdf of $X_i$ is $$f(X=x)=\frac{1}{\theta} \exp\left( -\frac{x}{\theta} \right)$$  

Now let's check how [LLN](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln)
behaves for Exponential Distribution, $\text{Exp}(\theta)$  
Our expectation is $$\left|\overline{X}_n -\mu \right| \xrightarrow [n\to \infty ]{\mathbb{P},\text{ a.s.}} 0$$

```julia
 = 1
n = 300
"""   Exp(); pdf: f(=x) = (1/)e^(-x/) """
distribution = Exponential()
lln(distribution, n)
```

### $X_i\sim \text{Ber}(p)$ 
Now assume that our **sample** $X_1,X_2,....,X_n$ is drawn from a Bernoulli Distribution, $X_i\sim \text{Ber}(p)$  
pmf of $X_i$ is 

$$f(x,p)=  \left\{\begin{matrix}
p&; \text{if} &x=1\\
1-p&; \text{if} &x=0
\end{matrix}\right.$$

> in [Julia](https://juliastats.org/) Bernoulli Distribution is denoted as $\text{Bernoulli}(p)$

Now let's check how [LLN](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln)
behaves for Bernoulli Distribution, $\text{Ber}(p)$  
Our expectation is $$\left|\overline{X}_n -\mu \right| \xrightarrow [n\to \infty ]{\mathbb{P},\text{ a.s.}} 0$$

```julia
p = 0.5
n=300
distribution = Bernoulli(p)
lln(distribution, n)
```

### $X_i\sim \text{Geo}(p)$

> in [Julia](https://juliastats.org/) Geometric Distribution is denoted as $\text{Geometric}(p)$

Now let's check how [LLN](https://yuvraj97.github.io/stats.github.io/data/index/intro/./lln/#lln)

```julia
p = 0.5
n = 300
distribution = Geometric(p)
lln(distribution, n)
```
